```python
import nltk
import zipfile, os

# import reader
from nltk.corpus import PlaintextCorpusReader

# download library averaged_perceptron_tagger
nltk.download('averaged_perceptron_tagger')
# download library punkt
nltk.download('punkt')
```

```python
# unzip
# define filename
local_zip = 'brown.zip'
# open file with read mode
zip_ref = zipfile.ZipFile(local_zip, 'r')
# extract to tmp folder
zip_ref.extractall('tmp')
# close zip
zip_ref.close()
```

```python
# Inisialisasi struktur data
# make a map emit, transition, context
emit = {}
transition = {}
context = {}
```

```python
# Fungsi untuk memproses kalimat dan mengupdate emit, transition, dan context
def process_sentence(sentence):
    # previous = “<s>” # Make the sentence start
    previous = "<s>"
    #context[previous]++
    context[previous] = context.get(previous, 0) + 1
    # split line into wordtags with “ “
    # wordtags = line.strip().split(" ") #error tidak sesuai dataset
    # for each wordtag in wordtags
    for wordtag in sentence:
        #split wordtag into word, tag with “_”
        # word, tag = wordtag.split("_") #error tidak sesuai dataset
        word, tag = wordtag
        # transition[previous+“ “+tag]++ # Count the transition
        transition_key = f"{previous} {tag}"
        transition[transition_key] = transition.get(transition_key, 0) + 1
        # context[tag]++ # Count the context
        context[tag] = context.get(tag, 0) + 1
        # emit[tag+“ “+word]++ # Count the emission
        emit_key = f"{tag} {word}"
        emit[emit_key] = emit.get(emit_key, 0) + 1
        # previous = tag
        previous = tag
    # transition[previous+” </s>”]++
    transition_key = f"{previous} </s>"
    transition[transition_key] = transition.get(transition_key, 0) + 1
```

```python
# load all corpus file
corpus_root = '/content/tmp'
file_pattern = '.*'
brown_corpus = PlaintextCorpusReader(corpus_root, file_pattern)
```

```python
# proses data train (ca01 - ca40)
for file_id in brown_corpus.fileids()[:40]:
    sentences = brown_corpus.sents(file_id)
    for sentence in sentences:
        tagged_sentence = nltk.pos_tag(sentence)
        process_sentence(tagged_sentence)
```

```python
# Menghitung dan mencetak probabilitas transisi untuk data pelatihan
# Print the transition probabilities
# for each key, value in transition
for key, value in transition.items():
    # split key into previous, word with “ “
    previous, tag = key.split(" ")
    probability = value / context[previous]
    # print “T”, key, value/context[previous]
    print(f"T {key} {probability:.4f}")

# Do the same thing for emission probabilities with “E”
```

```python
# hitung probabilitas emisi untuk data training
for key, value in emit.items():
    tag, word = key.split(" ")
    probability = value / context[tag]
    print(f"E {key} {probability:.4f}")
```

```python
# Mengosongkan struktur data untuk data testing
emit.clear()
transition.clear()
context.clear()
```

```python
# Memproses data testing (ca41 - ca44)
for file_id in brown_corpus.fileids()[41:44]:
    sentences = brown_corpus.sents(file_id)
    for sentence in sentences:
        tagged_sentence = nltk.pos_tag(sentence)
        process_sentence(tagged_sentence)
```

```python
# Menghitung dan mencetak probabilitas transisi untuk data pengujian
for key, value in transition.items():
    previous, tag = key.split(" ")
    probability = value / context[previous]
    print(f"T {key} {probability:.4f}")
```

```python
# Menghitung dan mencetak probabilitas emisi untuk data pengujian
for key, value in emit.items():
    tag, word = key.split(" ")
    probability = value / context[tag]
    print(f"E {key} {probability:.4f}")
```
